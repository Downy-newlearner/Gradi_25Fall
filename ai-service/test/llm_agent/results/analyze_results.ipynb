```python
import pandas as pd

groq = pd.read_csv('results_meta-llama_llama-4-scout-17b-16e-instruct.csv')
```

```python
groq['response']
```

```python
groq[['answer', 'pred_answer']]
```

```python
# --------------------------
# 비교 함수
# --------------------------
def compare_answers(ans1, ans2):
    """두 답을 구조화 후 비교"""
    return ans1 == ans2

# --------------------------
# 적용
# --------------------------
groq['correct'] = groq.apply(lambda row: compare_answers(row['answer'], row['pred_answer']), axis=1)
```

```python
groq['correct'].value_counts()
```

```python
groq[groq['correct'] == False][['answer', 'pred_answer']].head(20)
```
